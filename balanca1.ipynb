```python
import serial

def criarSerial(portaSerial, baudRate): 
    try: 
        ser = serial.Serial(portaSerial, baudRate) 
        print("Conectado ao Arduino porta serial: " + portaSerial) 
        return ser 
    except: 
        print("Nao foi possivel a conexao com a porta serial:" + portaSerial)
```

```python
from datetime import datetime


def retornarHorario():
    atual = datetime.now()
    horario = atual.strftime("%H") + atual.strftime("%M") + atual.strftime("%f")
    return horario
```

```python
from datetime import datetime


def criarNome():
    horario = retornarHorario()
    nomeArq = horario +'ms'+".csv"
    return nomeArq
```

```python
def criarArquivo():
    nomeArq = criarNome()
    print("Arquivo criado")
    return nomeArq
```

```python
def abrirArquivo(nomeArq):
    caminhoArq = open(nomeArq, 'a')
    return caminhoArq
```

```python
def gravarArquivo(caminhoArq, dados):
    caminhoArq.write(dados + '; ' + "\n") #escreve os dados e pula de linha
```

```python
def fecharArquivo(caminhoArq):
    caminhoArq.close() #fecha o arquivo
```

```python
def fecharLeitura(ser):
    ser.close()
    print("Arquivo salvo e conexão serial encerrada")
```

```python
def imprimirOpcoes():
    print("-------------------------MENU-------------------------")
    print()
    print("- Para controlar a gravacao de dados, pressione:")
    print()
    print("p - para imprimir os dados")
    print("i - para interromper a aquisicao")
```

```python
from pynput.keyboard import Key

def verificarInput(key):
    try:
        print(key)
        if key.char == 'p':
            ser.write(b'p')
            print("Imprimindo valores por segundo")
        elif key.char == 'i':
            ser.write(b'i')
            print("Captura de eventos do teclado finalizada")
            return False #interrompe o listener
        else:
            print("Comando incorreto, tente novamente")
    except:
        print("Erro ao capturar comando")
```

```python
def retornarInicio():
    inicio = int(retornarHorario()[:2])
    if inicio < 4:
        return 1
    elif inicio >= 4 and inicio < 8:
        return 2
    elif inicio >= 8 and inicio < 12:
        return 3
    elif inicio >= 12 and inicio < 16:
        return 4
    elif inicio >= 16 and inicio < 20:
        return 5
    else:
        return 6
```

```python
from pynput.keyboard import Listener


def lerSerial(ser):
    ser.flushInput()
    nomeArq = criarArquivo()
    caminhoArq = abrirArquivo(nomeArq)
    imprimirOpcoes()
    escutador = Listener(on_press=verificarInput,on_release=None) #coleta eventos até pressionar esc
    escutador.start()
    flag = False
    cont = 0
    inicio = retornarInicio()
    while True:
        linha = str(ser.readline()) #.decode('utf-8')) #ler linha do terminal
        dados = linha[2:-5]
        gravarArquivo(caminhoArq, dados)
        #print(len(dados))
        cont = cont + 1
        if flag and cont >= 200:
            print(dados)
            cont = 0
        elif dados == "p":
            flag = not flag
        elif dados == "i":
            print("Escrita de arquivo interrompida")
            fecharArquivo(caminhoArq)#fecha o arquivo
            break
        else:
           if retornarInicio() != inicio:
                inicio = retornarInicio()
                fecharArquivo(caminhoArq)#fecha o arquivo de 4 em 4 horas
                nomeArq = criarNome()#cria um novo arquivo para continuar regitrando os dados
                caminhoArq = abrirArquivo(nomeArq)
    fecharLeitura(ser)
```

```python
ser = criarSerial("COM3","9600")
lerSerial(ser)
```
